```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import RadioButtons


def mandelbrot_set(x_min=-2, x_max=1, y_min=-1.5, y_max=1.5, n=1000, k=100):
    x = np.linspace(x_min, x_max, n)
    y = np.linspace(y_min, y_max, n) * 1j
    X, Y = np.meshgrid(x, y)
    c = X + Y
    z = np.zeros(c.shape)
    divergence_matrix = np.zeros(c.shape)
    for i in range(k):
        z = z ** 2 + c
        index = np.abs(z) < 2
        divergence_matrix[index] = i
    return divergence_matrix


fig, ax = plt.subplots()
plt.subplots_adjust(left=0.3, bottom=0.3)

divergence_matrix = mandelbrot_set()
im = plt.imshow(divergence_matrix, extent=[-2, 1, -1.5, 1.5], cmap='gray')

colors = ['jet', 'magma', 'viridis']

rax = plt.axes([0.025, 0.4, 0.15, 0.15], facecolor='lightgray')
radio = RadioButtons(rax, colors)


def change_colormap(label):
    im.set_cmap(label)
    fig.canvas.draw()


radio.on_clicked(change_colormap)

plt.xlim([-2, 1])
plt.ylim([-1.5, 1.5])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button

k = 100

# Create the figure and the line that we will manipulate
fig, ax = plt.subplots()


# adjust the main plot to make room for the sliders
fig.subplots_adjust(left=0.25, bottom=0.25)


divergence_matrix = mandelbrot_set()
im = plt.imshow(divergence_matrix, extent=[-2, 1, -1.5, 1.5], cmap='gray')


# Make a vertically oriented slider to control the amplitude
axfreq = fig.add_axes([0.25, 0.1, 0.65, 0.03])
freq_slider = Slider(
    ax=axfreq,
    label='Frequency [Hz]',
    valmin=0,
    valmax=200,
    valinit=k,
)


# The function to be called anytime a slider's value changes
def update(val):
    
    # Get the new k value
    k = int(freq_slider.val)
    
    # Generate the new Mandelbrot set
    divergence_matrix = mandelbrot_set()
    
    # Update the plot with the new Mandelbrot set
    im.set_data(divergence_matrix)
    
    # Redraw the plot
    fig.canvas.draw_idle()


# register the update function with each slider

freq_slider.on_changed(update)


plt.show()
```
